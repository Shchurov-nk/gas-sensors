```python
import pandas as pd
import plotly.express as px
```

```python
def plot_corr_xx(corr_xx):
    # corr_xx.columns = corr_xx.columns.astype(float)
    fig = px.imshow(
        corr_xx.iloc[0::5, 0::5], 
        color_continuous_scale='YlOrRd', 
        labels={'x': "Wavenumber, cm⁻¹"})
    fig.update_layout(
        xaxis={'side': 'top'}, 
        yaxis={'side': 'left'}
        )
    return fig
```

```python
sensor_name = 'R12'
corr_xx = pd.read_feather(f'../data/corr/XX/corr_XX-{sensor_name}.feather')
corr_xy = pd.read_feather(f'../data/corr/XY/corr_XY-{sensor_name}.feather')
```

```python
corr_xx[['R12_0']].plot()
corr_xy['NO_conc'].plot()
```

```python
plot_corr_xx(corr_xx)
```
